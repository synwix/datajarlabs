```python
import numpy as np
import pandas as pd
from sklearn import linear_model
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings('ignore')
train = pd.read_csv('./train.csv')
train.head(10)
```

```python
train.isnull().sum() *100 / train.shape[0]
```

```python
train["Age"] = train["Age"].interpolate()
del train["Cabin"]
```

```python
train.isnull().sum() *100 / train.shape[0]
```

```python
y = train["Survived"]
X = train[["Sex","Age","Pclass","Embarked"]]
X
```

```python
X= pd.concat([X ,pd.get_dummies(X["Sex"])],axis=1)

del X["Sex"]

X
```

```python
X = pd.concat([X ,pd.get_dummies(X["Embarked"])],axis=1)

del X["Embarked"]

X
```

```python
from sklearn.model_selection import train_test_split
X_eğitim, X_test, y_eğitim, y_test =  train_test_split(X, y, test_size=0.20, random_state=111)
```

```python
X
```

```python
from sklearn.linear_model import LogisticRegression
log_reg = LogisticRegression()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test =  train_test_split(X, y, test_size=0.20, random_state=111)
```

```python
log_reg.fit(X_train, y_train)
```

```python
train_accuracy = log_reg.score(X_train, y_train)
test_accuracy = log_reg.score(X_test, y_test)
print('One-vs-rest', '-'*20, 
      'Modelin eğitim verisindeki doğruluğu : {:.2f}'.format(test_accuracy), 
      'Modelin test verisindeki doğruluğu   : {:.2f}'.format(test_accuracy), sep='\n')
```

```python
log_reg_mnm = LogisticRegression(multi_class='multinomial', solver='lbfgs')
log_reg_mnm.fit(X_eğitim, y_eğitim)
train_accuracy = log_reg_mnm.score(X_train, y_train)
test_accuracy = log_reg_mnm.score(X_test, y_test)
print('Multinomial (Softmax)', '-'*20, 
      'Modelin eğitim verisindeki doğruluğu : {:.2f}'.format(test_accuracy), 
      'Modelin test verisindeki doğruluğu   : {:.2f}'.format(test_accuracy), sep='\n')
```
